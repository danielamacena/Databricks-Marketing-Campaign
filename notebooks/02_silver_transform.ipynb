```python
from pyspark.sql.functions import col

df_bronze = spark.table("bronze_marketing_campaigns")

df_silver = df_bronze \
    .filter(col("date").isNotNull()) \
    .filter(col("campaign_id").isNotNull()) \
    .filter(col("campaign_name").isNotNull()) \
    .filter(col("impressions") > 0) \
    .filter(col("clicks") > 0) \
    .filter(col("conversions") >= 0) \
    .filter(col("cost") > 0) \
    .filter(col("revenue") >= 0)

df_silver.write.format("delta") \
    .mode("overwrite") \
    .saveAsTable("silver_marketing_campaigns")
```
