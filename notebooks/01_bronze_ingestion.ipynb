```python
from delta.tables import DeltaTable
from pyspark.sql.functions import col

source_df = spark.table("workspace.default.marketing_campaign_data")

source_df.createOrReplaceTempView("source_marketing_campaign_data")

spark.sql("""
MERGE INTO bronze_marketing_campaigns AS target
USING source_marketing_campaign_data AS source
ON target.date = source.date
AND target.campaign_id = source.campaign_id
AND target.region = source.region

WHEN MATCHED THEN UPDATE SET
  target.campaign_name = source.campaign_name,
  target.channel = source.channel,
  target.impressions = source.impressions,
  target.clicks = source.clicks,
  target.conversions = source.conversions,
  target.cost = source.cost,
  target.revenue = source.revenue,
  target.last_updated = source.last_updated

WHEN NOT MATCHED THEN INSERT *
""")
```
